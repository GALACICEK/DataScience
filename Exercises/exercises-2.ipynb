```python
def make_frame(country:str):
    """Take country name as parameter,
    return inflected,recovered, deaths values as timeseries
    """
    df = pd.DataFrame({
        'inflected' : infected_country.loc[country],
        'recovered' : recovered_country.loc[country],
        'deaths' : deaths_country.loc[country]
    })

    df.index = pd.to_datetime(df.index, format='%m/%d/%y')
    return df
```

```python
df_us = make_frame('US')
df_turkey = make_frame('Turkey')
```

```python
#basic reproduction number  https://en.wikipedia.org/wiki/Basic_reproduction_number
#yayılım devam edip etmemesini kontrol ediyoruz
df_us['Rt'] = df_us['new_inflected'].rolling(8).apply(lambda x: x[4:].sum() / x[:4].sum() )
#apply methodunun içine for döngüsü ya da koşullar eklenebilir
#arada veri eksiği aolabilir ya da infinite değerler çıkıyor olabilir
df_us
```

```python
df_us['Rt'].plot()
plt.show()
```

```python
""" na ve inf değerler için neler yapılabilir
    log değerler alınabilir
    eksik verileri tamamlamak
    eksik verileri doldurmak için pandas methodlarına bakmak
"""
```

```python
df_us['Rt'].replace(np.inf, np.nan).fillna(method='bfill')
```

```python
rt_array = df_us[df_us.index < '2020-06-01']['Rt'].replace(np.inf, np.nan).bfill()
print(rt_array)
```
