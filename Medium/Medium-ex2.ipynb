```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Örnek veri seti oluşturalım (örneğin, aylara göre satış verileri)
aylar = ['Ocak', 'Şubat', 'Mart', 'Nisan', 'Mayıs', 'Haziran']

# Rastgele satış verileri oluşturmak için NumPy kullanalım
np.random.seed(42)  # Her seferde aynı veriyi elde etmek için seed kullanıyoruz
satis_verileri = np.random.randint(100, 300, size=6)  # 100 ile 300 arasında rastgele 6 sayı

# Çizgi grafiğini oluşturmak için Matplotlib kullanalım
""" Bu linklerdeki versiyonlarlakullandığınız versiyonları göz önüne alıp inceleyin.
    plt.plot(xekseni,yekseni, marker = "https://matplotlib.org/stable/api/markers_api.html",
            linestyle = https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html
            color = https://matplotlib.org/stable/gallery/color/colormap_reference.html#sphx-glr-gallery-color-colormap-reference-py)
"""
plt.plot(aylar, satis_verileri, marker='X', linestyle=':', color='b', label='Satışlar')

# Grafik başlığı ve ekseni etiketlerini ekleyelim
plt.title('Aylara Göre Satış Grafiği')
plt.xlabel('Aylar')
plt.ylabel('Satış Miktarı')

# Grafik üzerindeki çizgiyi ve noktaları gösterelim
"""plt.legend() kullanarak grafik üzerinde her bir çizgi veya veri seti için bir açıklama ekleyebiliriz. 
   Bu açıklamalar genellikle grafiğin sağ üst köşesine yerleştirilir,  konumunu özelleştirmek de mümkündür.
"""
plt.legend()

# Grafiği gösterelim
plt.show()
```

```python
# Örnek veri seti
kategoriler = ['Kategori A', 'Kategori B', 'Kategori C', 'Kategori D']
np.random.seed(35)  # Her seferde aynı veriyi elde etmek için seed kullanıyoruz
veri_seti  = np.random.randint(20, 100, size=4)  # 20 ile 100 arasında rastgele 4 sayı


# Sütun grafiklerini oluşturmak için Matplotlib kullanalım
plt.bar(kategoriler, veri_seti, color='teal')

# Eksen etiketleri ve başlığı ekleyelim
plt.xlabel('Kategoriler')
plt.ylabel('Frekans/Değer')
plt.title('Bar Chart Örneği')

# Sütunların üzerine değerleri ekleyelim
for i, v in enumerate(veri_seti):
    plt.text(i, v + 1, str(v), ha='center', va='bottom')

# Grafikleri gösterelim
plt.show()
```

```python
kategoriler = ['Kategori A', 'Kategori B', 'Kategori C', 'Kategori D']
veri_seti = [30, 45, 12, 55]

# Pasta grafiğini oluşturmak için Matplotlib kullanalım
plt.pie(veri_seti, labels=kategoriler, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen', 'lightcoral', 'lightgoldenrodyellow'])

# Pasta grafiğinin çember şeklinde olmasını sağlayalım
plt.axis('equal')

# Başlık ekleyelim
plt.title('Pasta Grafik Örneği')

# Grafiği gösterelim
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Örnek veri setini oluşturalım
veri = {
    'Kategori': ['A', 'A', 'B', 'B', 'C', 'C'],
    'Değer': [4, 7, 2, 5, 8, 3],
    'Frekans': [10, 15, 8, 12, 9, 7]
}

df = pd.DataFrame(veri)

# Pivot tablo oluşturalım
pivot_tablo = df.pivot_table(values='Frekans', index='Kategori', columns='Değer', fill_value=0)

plt.figure(figsize=(10,10))
# Heat map oluşturalım
plt.imshow(pivot_tablo, cmap='YlGnBu', interpolation='nearest', aspect='auto')

# Renk çubuğu ekleyelim
plt.colorbar(label='Frekans')

# Eksen etiketlerini ve başlığı ekleyelim
plt.xticks(np.arange(len(pivot_tablo.columns)), pivot_tablo.columns)
plt.yticks(np.arange(len(pivot_tablo.index)), pivot_tablo.index)
plt.xlabel('Değer')
plt.ylabel('Kategori')
plt.title('Heat Map Örneği')

# Grafiği gösterelim
plt.show()
```

```python
# Örnek veri setini oluşturalım
veri = {
    'Kategori': ['A', 'A', 'B', 'B', 'C', 'C'],
    'Değer': [4, 7, 2, 5, 8, 3],
    'Frekans': [10, 15, 8, 12, 9, 7]
}

df = pd.DataFrame(veri)

# Pivot tablo oluşturalım
pivot_tablo = df.pivot_table(values='Frekans', index='Kategori', columns='Değer', fill_value=0)

# Çizilecek figürün boyutlarını ayarlayalım
plt.figure(figsize=(8,8))

# Heat map oluşturalım
plt.imshow(np.triu(pivot_tablo.values), cmap='YlGnBu', interpolation='nearest', aspect='auto')

# Renk çubuğu ekleyelim
plt.colorbar(label='Frekans')

# Eksen etiketlerini ve başlığı ekleyelim
plt.xticks(np.arange(len(pivot_tablo.columns)), pivot_tablo.columns)
plt.yticks(np.arange(len(pivot_tablo.index)), pivot_tablo.index)
plt.xlabel('Değer')
plt.ylabel('Kategori')
plt.title('Üst Üçgen Heat Map Örneği')

# Grafiği gösterelim
plt.show()
```

```python
# Örnek veri setini oluşturalım
veri = {
    'X': np.random.rand(150),  # Rastgele X değerleri
    'Y1': 3 * np.random.rand(150) + 2, # Y1 = 3X + 2 + gürültü
    'Y2': 5 * np.random.rand(150) + 3  #  Y2 = 5X + 3 + gürültü
}

df = pd.DataFrame(veri)

# Dağılım grafiğini oluşturalım
plt.scatter(df['X'], df['Y1'], color='cyan', alpha=0.7, edgecolors='w', label='Y1')
plt.scatter(df['X'], df['Y2'], color='firebrick', alpha=0.7, edgecolors='w', label='Y2')

# Eksen etiketleri ve başlığı ekleyelim
plt.xlabel('X Değeri')
plt.ylabel('Y Değeri')
plt.title('İki Değişkenli Dağılım Grafiği Örneği')

# Legend (açıklama) ekleyelim
plt.legend()

# Grafikleri gösterelim
plt.show()
```

```python
# Örnek veri setini oluşturalım
veri = {
    'Değerler': np.concatenate([
        np.random.randn(200) * 5 + 30,
        np.random.randn(200) * 8 + 50,
        np.random.randn(200) * 4 + 40,
        np.random.randn(200) * 6 + 45,
        np.random.randn(200) * 7 + 55
    ])
}

df = pd.DataFrame(veri)

# Histogramı oluşturalım
plt.hist(df['Değerler'], bins=30, color='skyblue', edgecolor='black')

# Eksen etiketleri ve başlığı ekleyelim
plt.xlabel('Değerler')
plt.ylabel('Frekans')
plt.title('Histogram Örneği - 5 Farklı Değer')

# Grafikleri gösterelim
plt.show()
```

```python
# 5 farklı veri setini oluşturalım
veri_set1 = np.random.normal(loc=50, scale=10, size=100)
veri_set2 = np.random.normal(loc=45, scale=8, size=100)
veri_set3 = np.random.normal(loc=55, scale=12, size=100)
veri_set4 = np.random.normal(loc=40, scale=15, size=100)
veri_set5 = np.random.normal(loc=60, scale=10, size=100)

veri = {'Veri Seti 1': veri_set1, 'Veri Seti 2': veri_set2, 'Veri Seti 3': veri_set3,
        'Veri Seti 4': veri_set4, 'Veri Seti 5': veri_set5}

df = pd.DataFrame(veri)

# Kutu grafikleri oluşturalım
plt.boxplot(df.values, labels=df.columns)

# Eksen etiketleri ve başlığı ekleyelim
plt.xlabel('Veri Setleri')
plt.ylabel('Değerler')
plt.title('Kutu Grafikleri - 5 Farklı Veri Seti')

# Grafikleri gösterelim
plt.show()
```
